```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
rng = np.random.default_rng(seed=80)

positions = rng.integers(0, 3, size=(2_000, 2))
positions = np.where(positions > 1, -1, positions)
result = positions.cumsum(axis=0)
```

```python
fig = plt.figure()
plt.title("Brownian Motion 2D Simulation - 2000 timepoints");
axMax = max(max(abs(result[:, 0].min()), abs(result[:, 0].max())), max(abs(result[:, 1].min()), abs(result[:, 1].max())))
plt.xlim((-1 * axMax) - 10, axMax + 10);
plt.ylim((-1 * axMax) - 10, axMax + 10);
graph, = plt.plot([], [], '-b')

def animate(i):
    graph.set_data(result[:i + 1, 0], result[:i + 1, 1])
    return graph

ani = FuncAnimation(fig, animate, frames=2_000, interval=20)

from IPython.display import HTML
HTML(ani.to_html5_video())
```

```python
# furthest point from origin

result[np.argmax(distances)]
```
